```python
import pandas as pd
import time
import requests
from ratelimit import limits, sleep_and_retry
RATE_LIMIT = 5000
RATE_LIMIT_WINDOW = 3600  # 1 hour in seconds
```

```python
headers = {'Authorization': 'token give_your_token'}
df=pd.read_csv('F:\\NeubAItics Tech Pvt Ltd\\data\\datasplit\\file4.csv')
```

```python
@sleep_and_retry
@limits(calls=RATE_LIMIT, period=RATE_LIMIT_WINDOW)
def make_request(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 403:
        time.sleep(3600)  
        print("Sleeping time")
        response = requests.get(url, headers=headers)  
    response.raise_for_status()
    return response
def get_commits_count(user_name, repo_name):
    """Get the count of commits made by a specific author to a GitHub repository."""
    commits_count = 0
    url = f'https://api.github.com/repos/{user_name}/{repo_name}/commits'
    params = {'author': user_name, 'per_page': 100}

    while url:
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            commits = response.json()
            commits_count += len(commits)
            if 'next' in response.links:
                url = response.links['next']['url']
            else:
                break
        else:
            print(f"Failed to fetch commits. Status code: {response.status_code}")
            break

    return commits_count
def get_user_details(username):
    """Fetch a user's public details from GitHub."""
    url = f'https://api.github.com/users/{username}'
    response = make_request(url)  
    if response.status_code == 200:
        user_details = response.json()
        return user_details
    elif response.status_code == 403:
        print("Rate limit exceeded. Waiting before retrying...")
        time.sleep(60)  
        return get_user_details(username) 
    else:
        print(f"Failed to fetch user details. Status code: {response.status_code}")
        return None
def get_repos(user):
    """Fetch a list of repositories for a given user."""
    repos_url = f'https://api.github.com/users/{user}/repos'
    response = requests.get(repos_url, headers=headers)
    repos = response.json()
    return [repo['name'] for repo in repos]

def get_languages_for_repo(user, repo):
    """Fetch programming languages used in a given repository."""
    languages_url = f'https://api.github.com/repos/{user}/{repo}/languages'
    response = requests.get(languages_url, headers=headers)
    return response.json()

def aggregate_languages(user):
    """Aggregate programming languages across all repositories of a user."""
    repos = get_repos(user)
    aggregated_languages = {}

    for repo in repos:
        languages = get_languages_for_repo(user, repo)
        for language, bytes_used in languages.items():
            if language in aggregated_languages:
                aggregated_languages[language] += bytes_used
            else:
                aggregated_languages[language] = bytes_used

    return aggregated_languages

def get_orgs(user):
   """Get all organisations for a user."""
   orgs_url = f'https://api.github.com/users/{user}/orgs'
   response = requests.get(orgs_url, headers=headers)
   return response.json()

def get_subscriptions(user):
   """Get all subscriptions for a user."""
   subscriptions_url = f'https://api.github.com/users/{user}/subscriptions'
   response = requests.get(subscriptions_url, headers=headers)
   return response.json()

def get_starred_repo(user):
   """Get all starred repo for a user."""
   starred_repo_url = f'https://api.github.com/users/{user}/starred'
   response = requests.get(starred_repo_url, headers=headers)
   return response.json()


def get_follower_list(user):
   """Get all follower list for a user."""
   follower_url = f'https://api.github.com/users/{user}/followers'
   response = requests.get(follower_url, headers=headers)
   return response.json()

def get_following_list(user):
   """Get all following list for a user."""
   following_url = f'https://api.github.com/users/{user}/following'
   response = requests.get(following_url, headers=headers)
   return response.json()
```

```python
# Authenticate with your personal access token
user_data = []
for each_username in df["username"]:
    try:
        each_user_detail = {}
        user = get_user_details(each_username)
        all_user_repos = get_repos(each_username)
        # Fetch the required information
        total_commits = sum([get_commits_count(each_username, repo) for repo in all_user_repos])
        languages = aggregate_languages(each_username)
        each_user_detail["Full_Name"] = user['name']
        each_user_detail["Starred Repositories"] = [each_repo['url'] for each_repo in get_starred_repo(each_username)]
        each_user_detail["Subscriptions"] = [each_repo['url'] for each_repo in get_subscriptions(each_username)]
        each_user_detail["Organizations"] = [each_repo['login'] for each_repo in get_orgs(each_username)]
        each_user_detail["Login"] = user["login"]
        each_user_detail["Bio"] = user["bio"]
        each_user_detail["Public Repositories"] = user["public_repos"]
        each_user_detail["Followers Count"] = user["followers"]
        each_user_detail["Following Count"] = user["following"]
        each_user_detail["Created At"] = user["created_at"]
        each_user_detail["Updated At"] = user["updated_at"]
        each_user_detail["Avatar URL"] = user["avatar_url"]
        each_user_detail["Languages"] = languages
        each_user_detail["Profile URL"] = user["url"]
        each_user_detail["Total Commits"] = total_commits
        each_user_detail["Followers List"] = [each_user["login"] for each_user in get_follower_list(each_username)]
        each_user_detail["Following List"] = [each_user["login"] for each_user in get_following_list(each_username)]
        user_data.append(each_user_detail)
        print(each_username)
        time.sleep(1)  # Pause between requests to avoid exceeding the rate limit

    except Exception as e:
        print(f"Failed to fetch data for user {each_username}: {str(e)}")
```

```python
df=pd.DataFrame(user_data)
df.to_csv('file4_converted.csv', index=False)
print("data successfully converted---")
```
